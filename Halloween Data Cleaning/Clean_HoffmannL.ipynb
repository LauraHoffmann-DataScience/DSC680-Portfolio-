### Grouping with dicts and series

```python
# These lines will convert the degrees of separation into whole numbers and validate that they are numbers, or NaN values
candydata = candydata.replace('3 or higher', '3')
candydata['Please estimate the degree(s) of separation you have from the following celebrities [JK Rowling]'] = (pd.to_numeric(candydata['Please estimate the degree(s) of separation you have from the following celebrities [JK Rowling]'], errors='coerce'))
candydata['Please estimate the degree(s) of separation you have from the following celebrities [JJ Abrams]'] = pd.to_numeric(candydata['Please estimate the degree(s) of separation you have from the following celebrities [JJ Abrams]'], errors='coerce')
candydata['Please estimate the degree(s) of separation you have from the following celebrities [Beyoncé]'] = pd.to_numeric(candydata['Please estimate the degree(s) of separation you have from the following celebrities [Beyoncé]'], errors='coerce')
candydata['Please estimate the degree(s) of separation you have from the following celebrities [Bieber]'] = pd.to_numeric(candydata['Please estimate the degree(s) of separation you have from the following celebrities [Bieber]'], errors='coerce')
candydata['Please estimate the degree(s) of separation you have from the following celebrities [Kevin Bacon]'] = pd.to_numeric(candydata['Please estimate the degree(s) of separation you have from the following celebrities [Kevin Bacon]'], errors='coerce')
# Turns NaN values into the mode of the column
candydata=candydata.fillna(candydata.mode().iloc[0])
```

```python
# These lines will set all of the celebrity related variables equal to 'Degrees' for the groupby function
mapping = {'Please estimate the degree(s) of separation you have from the following celebrities [JK Rowling]':'Degrees',
          'Please estimate the degree(s) of separation you have from the following celebrities [JJ Abrams]':'Degrees',
          'Please estimate the degree(s) of separation you have from the following celebrities [Beyoncé]':'Degrees',
          'Please estimate the degree(s) of separation you have from the following celebrities [Bieber]':'Degrees',
          'Please estimate the degree(s) of separation you have from the following celebrities [Kevin Bacon]':'Degrees',
          'How old are you?':'Age'}
bycol = candydata.groupby(mapping, axis=1)
# This line will add all of the degrees away from celebrities each person is and output the total, the lower the number
# the closer one is to celebrities (I realize the number won't go higher than 15 because the max number of degrees away was
# 3 and there were 5 celebrities). The output will also include age although there isn't anything to add that to so the sum
# will just be their age and nothing else.
bycol.sum()
```

### Convert timestamps to periods and back

```python
# Import libraries and sets timestamp column equal to 'dates'
from datetime import datetime
import numpy as np

dates = candydata['Timestamp']
```

```python
# Sets the series using the dates and times from the timestamp column of the data with random numpy numbers
ts = pd.Series(np.random.randn(6889), index=dates)
ts
```

```python
# Converting the timestamps to periods
# Periods by hours of each date
periodts = ts.to_period(freq='H')
periodts
```

```python
# Converting the periods back to timestamps
# Using the "how='end'" means that it will add the last possible stamp for the minute and second range of the hour
periodts.to_timestamp(how='end')
```

### Generate a date range

```python
# This line will take the first timestamp from the candydata and create a range of timestamps
# one timestamp for every 2 hours dictated by the freq 2H and 101 timestamps from the periods
# adding normalize makes the hour stamps even on the dot starting at midnight with the first
# and adding two hours to every time stamp. Without normalize the same exact time would be 
# carried over for every 2 hours starting with the first
rng = pd.date_range(candydata['Timestamp'].min(), freq='2H', periods=101, normalize=True)
rng
```
